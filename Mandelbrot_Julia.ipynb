```julia
function mandelbrot(a)
    z = 0
    for i in 1:50
        if abs(z) > 1000
            return 4
        end
        z = z^2 + a
    end
    return z
end

function get_mandel()
    arr = zeros(3000, 3000)
    for (j,y) in enumerate(linspace(1., -1., 3000))
        for (i,x) in enumerate(linspace(-2., 0.5, 3000))
            z = abs(mandelbrot(complex(x,y)))
            if (z < 2)
                arr[j,i] = z
            else
                arr[j,i] = 2
            end
        end
    end
    return arr
end

@time arr = get_mandel()
```
