```julia
function mandelbrot(a)
    z = complex(0., 0.)
    for i in 1:50
        if abs(z) > 1000
            return i
        end
        z = z^2 + a
    end
    return 50
end

function get_mandel()
    arr = zeros(3000, 3000)
    for (j,y) in enumerate(linspace(1., -1., 3000))
        for (i,x) in enumerate(linspace(-2., 0.5, 3000))
            arr[j,i] = mandelbrot(complex(x,y))
        end
    end
    return arr
end
```

```julia
@time arr = get_mandel();
```
