```python
import numpy as np
from numba import jit

@jit
def mandelbrot(a):
    z = 0
    for i in range(51):
        if abs(z) > 1000:
            return i
        z = pow(z,2) + a
    return 50

@jit
def get_mandel():
    ar = np.zeros((3000,3000))
    for j, y in enumerate(np.linspace(1., -1., 3000)):
        for i, x in enumerate(np.linspace(-2., .5, 3000)):
            ar[j,i] = mandelbrot(complex(x, y))
    return ar
```

```python
%time arr = get_mandel()
```

```python
import matplotlib.pyplot as plt

plt.imshow(arr)
plt.show()
```
