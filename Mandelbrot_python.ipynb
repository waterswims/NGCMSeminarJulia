```python
import numpy as np

def mandelbrot(a):
    z = 0
    for i in range(51):
        if abs(z) > 1000:
            return 4
        z = pow(z,2) + a
    return z

def get_mandel():
    ar = np.zeros((3000,3000))
    for j, y in enumerate(np.linspace(1., -1., 3000)):
        for i, x in enumerate(np.linspace(-2., .5, 3000)):
            z = abs(mandelbrot(complex(x, y)))
            if (z < 2):
                ar[j,i] = z
            else:
                ar[j,i] = 2
    return ar

%timeit arr = get_mandel()
```

```python
import matplotlib.pyplot as plt

plt.imshow(arr)
plt.show()
```
